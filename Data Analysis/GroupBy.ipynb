```python
ps3 = games.loc[games.Platform == 'PS3',['Name','Genre','EU_Sales','Global_Sales']]
```

```python
ps3
```

```python
ps3.groupby('Genre').apply(lambda x: 'solid' if x.EU_Sales.sum() > 50 else 'weak')
```

```python
def stats_gm(i):
    levels = 'solid' if i.EU_Sales.sum() > 50 else 'weak'
    variability = 'volatile' if i.EU_Sales.std()/i.EU_Sales.mean() > 2 else 'steady'
    return variability,levels+' sales'
```

```python
ps3.groupby('Genre').apply(stats_gm)
```
