```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a = np.linspace(0,10.11)
```

```python
b= a**4
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
axes2.plot(x,y);
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([1,1,2,2])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,30)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in');
```

```python
fig1 = plt.figure(figsize =(5,5),dpi=300)
axes = fig1.add_axes([0,0,1,1])
axes.plot(a,b);
fig1.savefig('new_plot.png',bbox_inches = 'tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**3,label = 'X vs X^3')
ax.legend(loc = 0);
```
