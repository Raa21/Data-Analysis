```python
sat = pd.read_csv("C:/Users/rajen/Data Analysis/scores.csv")
```

```python
sat.head()
```

```python
math = sat[sat['SAT Section'] == 'Math'].Score.reset_index(drop=True)
read = sat[sat['SAT Section'] == 'Reading'].Score.reset_index(drop=True)
write = sat[sat['SAT Section'] == 'Writing'].Score.reset_index(drop=True)
```

```python
df = pd.concat([math,read,write],axis=1)
df.columns = ['Math','Read','Write']
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='Math',y='Write');
```

```python
math.corr(write)
```

```python
df.plot(kind='scatter',x='Read',y='Write');
```

```python
read.corr(write)
```
