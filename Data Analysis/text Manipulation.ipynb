```python
boston['Official Time'].str.replace(r'(\d+):(\d+):(\d+)',r'\1 hours, \2 minutes and \3 seconds',regex=True)
```

```python
final_times = boston['Official Time'].str.split(':',expand=True)
```

```python
final_times.rename({0:'Hours',1:'Minutes',2:'Seconds'},axis=1,inplace=True)
```

```python
final_times['Total'] = (final_times.Hours.astype(float)*3600)+(final_times.Minutes.astype(float)*60)+(final_times.Seconds.astype(float))
```

```python
final_times.head()
```

```python
boston.head()
```

```python
boston.insert(boston.columns.get_loc('Official Time'),'Total',final_times.Total)
```
